```python
from dataset.stations import *
import networkx as nx
import matplotlib.pyplot as plt
pos = {
    Tajrish: (0, 0),
    Shariati: (1, 0),
    Sattari: (0, 2),
    Sanat: (0, 1),
    Valiasr: (1, 1),
    Jahad: (2, 2),
    Beheshti: (2, 0),
    Azadi: (3, 0),
    Theather: (2, 1),
    Ferdowsi: (2, 3),
    Dowlat: (3, 0),
    Mohamadieh: (3, 1),
    Kahrizak: (5, 0),
    Molavi: (4, 1),
    Basij: (5, 1),
    Heravi: (4, 2),
    Ghaem: (5, 2),
    Shemiran: (4, 3),
    Booali: (5, 3)
}
metro_structure = {
    Kahrizak: [],
    Basij: [],
    Ghaem: [],
    Booali: [],
    Mohamadieh: [Kahrizak, Molavi],
    Dowlat: [Mohamadieh],
    Beheshti: [Dowlat, Heravi],
    Shariati: [Beheshti],
    Tajrish: [Shariati],
    Molavi: [Basij],
    Theather: [Mohamadieh, Ferdowsi],
    Valiasr: [Theather, Jahad],
    Sanat: [Valiasr],
    Heravi: [Ghaem],
    Jahad: [Beheshti],
    Sattari: [Valiasr],
    Shemiran: [Booali],
    Ferdowsi: [Shemiran],
    Azadi: [Theather],
}
# Create a directed graph from the metro structure
G = nx.DiGraph(metro_structure)
color_map = []
common_station = [Theather, Mohamadieh, Beheshti, Dowlat, Jahad, Valiasr]
for node in G:
    if node in common_station:
        color_map.append("brown")
    elif node in Line1:
        color_map.append("blue")
    elif node in Line2:
        color_map.append("red")
    elif node in Line3:
        color_map.append("green")
    else:
        color_map.append("yellow")

# pos = nx.kamada_kawai_layout(G, scale=1)
plt.figure(figsize=(18, 10))
nx.draw(G, pos, with_labels=True, node_size=2500, node_color=color_map, font_size=10, font_weight='bold', arrowstyle='-|>', arrowsize=12)
plt.title('Metro Station Structure (Improved Layout)')
plt.axis('off')  # Turn off the axis for a cleaner look
plt.show()
```
