```python
from dataset.stations import *
import networkx as nx
import matplotlib.pyplot as plt
metro_structure_line1 = {
    Kahrizak: [],
        Mohamadieh: [Kahrizak, Molavi],
    Dowlat: [Mohamadieh],
    Beheshti: [Dowlat, Heravi],
    Shariati: [Beheshti],
    Tajrish: [Shariati],

}
metro_structure_line2 = {
    Basij: [],
        Molavi: [Basij],
    Mohamadieh: [Kahrizak, Molavi],
    Theather: [Mohamadieh, Ferdowsi],
    Valiasr: [Theather, Jahad],
    Sanat: [Valiasr],
}
metro_structure_line3 = {
    Ghaem: [],
        Heravi: [Ghaem],
        Beheshti: [Dowlat, Heravi],
        Jahad: [Beheshti],
        Valiasr: [Theather, Jahad],
        Sattari: [Valiasr],
}
metro_structure_line4 = {
    Booali: [],
        Shemiran: [Booali],
        Dowlat: [Mohamadieh],
        Ferdowsi: [Shemiran],
        Theather: [Mohamadieh, Ferdowsi],
        Azadi: [Theather],
}

edges_list = [(Booali, Shemiran)]
# Create a directed graph from the metro structure
G = nx.MultiDiGraph()
G.add_nodes_from(metro_structure_line1)
G.add_nodes_from(metro_structure_line2)
G.add_nodes_from(metro_structure_line3)
G.add_nodes_from(metro_structure_line4)
G.add_edges_from(edges_list)
pos = nx.kamada_kawai_layout(G, scale=1)
plt.figure(figsize=(18, 10))
nx.draw(G, pos, with_labels=True, node_size=2500, node_color='skyblue', font_size=10, font_weight='bold', arrowstyle='-|>', arrowsize=12)
plt.title('Metro Station Structure (Improved Layout)')
plt.axis('off')  # Turn off the axis for a cleaner look
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_metro_structure(metro_structure):
    # Create a directed graph from the metro structure
    G = nx.DiGraph(metro_structure)
    pos = nx.kamada_kawai_layout(G)

    plt.figure(figsize=(18, 10))
    nx.draw(G, pos, with_labels=True, node_size=2500, node_color='skyblue', font_size=10, font_weight='bold', arrowstyle='-|>', arrowsize=12,)
    plt.title('Metro Station Structure (Improved Layout)')
    plt.axis('off')  # Turn off the axis for a cleaner look
    plt.show()

# Usage:
metro_structure = {
    Kahrizak: [],
    Basij: [],
    Ghaem: [],
    Booali: [],
    Mohamadieh: [Kahrizak, Molavi],
    Dowlat: [Mohamadieh],
    Beheshti: [Dowlat, Heravi],
    Shariati: [Beheshti],
    Tajrish: [Shariati],
    Molavi: [Basij],
    Theather: [Mohamadieh, Ferdowsi],
    Valiasr: [Theather, Jahad],
    Sanat: [Valiasr],
    Heravi: [Ghaem],
    Jahad: [Beheshti],
    Sattari: [Valiasr],
    Shemiran: [Booali],
    Ferdowsi: [Shemiran],
    Azadi: [Theather],
}

draw_metro_structure(metro_structure)
```
